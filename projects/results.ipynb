```python
from metaflow import Flow
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
run = Flow('ClassifierTrainFlow').latest_run
run
```

```python
model, score = run.data.results[0]
test_data = run['start'].task.data.test_data
test_labels = run['start'].task.data.test_labels
```

Deprecation: The plot_confusion_matrix method has been deprecated and removed in newer versions of scikit-learn.

Replacement: Use ConfusionMatrixDisplay.from_estimator as the replacement.

```python
ConfusionMatrixDisplay.from_estimator(model, test_data, test_labels)
```
